```python
for numero in range(-100, 201):
    print(numero)




```


En este código usé la función range para generar una secuencia de números que va desde el -100 hasta 200. Luego,con la función for permití que se replicara la acción de print en todo numero que se encontrara en este intervalo. por último print permite que se vea toda la secuencia.

```python
for numero in range(-100, 201):
    if numero % 6 == 0:
        print(numero)
```

En un caso similar al primero, usé la función range para delimitar el rango y for para crear el bucle que para impirmir todis los numeros que cumplan las caracteristicas pedidas, solo que ahora usando if como condicionante, que en este caso era que la división diera resto 0, para lo que usé el símbolo % para que el programa entendiera lo que le pedía

```python
numero1 = 100
numero2 = 130

def sumar_numeros(numero1, numero2):
    resultado = numero1 + numero2
    if resultado < 200:
        print("Menor a 200")
    elif resultado <= 250:
        print("Mayor a 200")
    else:
        print("Mayor a 250")

sumar_numeros(numero1, numero2)

```

en este caso definí la función  sumar_numeros, que toma dos números como argumentos, los suma y luego imprime un mensaje según el resultado al final del codigo al usar print. en este caso se definieron distintas condicionantes con if, elif y else, para así darle las instrucciones al programa para que en los 3 distintos escenarios que me de la suma, definido previamente al darle un valor numérico a los datos numero1 y numero2, se imprimiera un mensaje distinto dada el resultado de la suma.

```python
edad_usuario = 22
gusta_musica_urbana = True

def evaluar_usuario(edad, gusta_musica_urbana):
    if edad >= 18 and gusta_musica_urbana:
        return "Eres mayor de edad y te gusta la música urbana. Bacán!"
    elif edad >= 18 and not gusta_musica_urbana:
        return "Eres mayor de edad, pero no te gusta la música urbana. Aburrido!"
    elif edad < 18 and gusta_musica_urbana:
        return "Eres menor de edad, pero te gusta la música urbana."


print(evaluar_usuario(edad_usuario, gusta_musica_urbana))
```

En este código, la función evaluar_usuario toma dos argumentos: edad (la edad del usuario) y gusta_musica_urbana (un boolean que indica si al usuario le gusta la música urbana). Luego, la función evalúa estas condiciones utilizando declaraciones if y elif,  dependiendo de si al usuario le gusta o no la música urbana y la edad del mismo, que para efectos practicos son valores definidos antes de crear la función, y devuelve un mensaje pre determinado según las respuestas del usuario.
